```python
for line in open('25cm_1.00034.lvm'):
    d=line.split()
    if len(d)!=4:
        continue
    data_25_t=np.append(data_25_t,float(d[0])); data_25_wn=np.append(data_25_wn,float(d[1]))
    data_25_r1=np.append(data_25_r1,float(d[2])); data_25_r2=np.append(data_25_r2,float(d[3]))

cc_25=signal.correlate(data_25_r2,data_25_r1)
lags = signal.correlation_lags(len(data_25_r1), len(data_25_r2))

max_cc_25=np.max(cc_25)
x_max_cc_25=lags[cc_25.argmax()]/(f_ac)
print(max_cc_25,x_max_cc_25)

plt.plot(lags/f_ac,cc_25)
plt.xlim(-0.01,0.01)
plt.show()
```

```python
D=np.linspace(0.25,0.95,15)
dt=np.array([x_max_cc_25,x_max_cc_30,x_max_cc_35,x_max_cc_40,x_max_cc_45,x_max_cc_50,x_max_cc_55,x_max_cc_60,
             x_max_cc_65,x_max_cc_70,x_max_cc_75,x_max_cc_80,x_max_cc_85,x_max_cc_90,x_max_cc_95])

error_distance=0.01

from iminuit import Minuit
from iminuit.cost import LeastSquares

def model(x, p):
    return x * p[0] + p[1]

fnc = LeastSquares(dt, D, error_distance, model)
m = Minuit(fnc, (343, 0))
m.migrad()
print(m)

plt.errorbar(dt,D,yerr=error_distance,color='blue')
plt.plot(dt,model(dt,m.values),color='r')
plt.show()
```
